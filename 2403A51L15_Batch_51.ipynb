# Aim :
#### To implement Linear Regression using Python and apply it to the Boston Housing Dataset to predict house prices.


```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error,mean_absolute_percentage_error
```

```python
# Load California housing dataset
california = fetch_california_housing()
data=pd.DataFrame(california.data,columns=california.feature_names)
```

```python
data['PRICE']=california.target
print(data.head())
```

```python
print(data.isnull().sum)
x=data.drop('PRICE',axis=1)
y=data['PRICE']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(data.isnull().sum)
x=data.drop('PRICE',axis=1)
y=data['PRICE']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
mse=mean_squared_error(y_test,model.predict(x_test))
mae=mean_absolute_error(y_test,model.predict(x_test))
mape=mean_absolute_percentage_error(y_test,model.predict(x_test))
print(f'MSE: {mse:.2f}')
print(f'MAE: {mae:.2f}')
print(f'MAPE: {mape:.2f}')
```

```python
plt.scatter(y_test,model.predict(x_test),color='grey',label='predit')
plt.scatter(y_test,y_test,color='black',label='actual')
plt.xlabel('Actual Prices ')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```
